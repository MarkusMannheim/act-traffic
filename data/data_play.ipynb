```python
import pandas as pd

areas = ["Belconnen", "Canberra", "City", "Gungahlin", "Tuggeranong", "Woden"]
cols = ["Location", "Direction", "Pre_Covid", "During_Covid", "Post_Covid", "Long", "Lat"]
columns = ["site", "dir", "pre", "dur", "post", "lng", "lat"]

df = pd.DataFrame(columns=columns + ["area"])

for area in areas:    
    df_len = len(df)
    filePath = f"{area}Centre_TrafficVolumes.xlsx" if area != "Canberra" else "CanberraRegion_TrafficVolumes.xlsx"
    data = pd.read_excel(filePath)
    
    def clean_loc(x):
        x = str(x).replace(", ", "").replace("  ", " ")
        if " (" in x:
            x = x[:x.index(" (")]        
        return x    
    data.Location = data.Location.apply(clean_loc)
    
    for i, indice in enumerate(data.index):
        for j, col in enumerate(cols):
            df.at[df_len + i, columns[j]] = data.at[i, col]
        df.at[df_len + i, "area"] = area

df.to_csv("./traffic_data.csv")
```
